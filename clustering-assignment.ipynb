```python
#create an array from labels and values
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

face = misc.face(gray=True)
```

```python
n_clusters = 5
np.random.seed(0)

L = face.reshape((-1,1))
k_means = KMeans(n_clusters= n_clusters,n_init=10)
k_means.fit(L)
vals = k_means.cluster_centers_.squeeze()
labels= k_means.labels_
```

```python
f_compressed = np.choose(labels,vals)
```

```python
f_compressed.shape = face.shape
```

```python
vmin = face.min()
vmax = face.max()
```

```python
#original face
plt.figure(1)
plt.imshow(face, cmap=plt.cm.gray, vmin=vmin, vmax=256)
```

```python
#compressed face
plt.figure(2)
plt.imshow(f_compressed, cmap=plt.cm.gray, vmin=vmin, vmax=vmax)
```
